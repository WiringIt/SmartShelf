```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
y=np.empty((250,3))
for i in range(1,4):
  df=pd.read_csv('0{}.csv'.format(i))
  df = df.drop(df.columns[-1], axis=1)
  for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
  medians=df.median()
  for j in range(0,250):
    y[j][i-1]=medians.iloc[j]
#print(medians)
# print(medians[0:5])
# print(medians.shape)
x=np.empty(250,)
k=0
for i in range(50000,300000,1000):
  x[k]=i
  k+=1
print(x)
print(x.shape)
print("************")
print(y[:,0].shape)
plt.plot(x,y[:,0],linestyle='-',marker='.',linewidth=1,label='0Iteration1')
plt.plot(x,y[:,1],linestyle='-.',marker='*',linewidth=2,label='0Iteration2')
plt.plot(x,y[:,2],linestyle=':',marker='^',linewidth=3,label='0Iteration3')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
df.head()
```

```python
#normalising the dataset for cell 1 across iterations
df01=pd.read_csv('01.csv')
df1=pd.read_csv('11.csv')
for col in df1.columns:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
df1.head()
for col in df01.columns:
    df01[col] = pd.to_numeric(df01[col], errors='coerce')
df01.head()

for i in range(0,250):
     for j in range(0,500):
      #print("value in 1",df1.iloc[j,i])
      #print("value in ref",df0.iloc[j,i])
      df1.iloc[j,i]=df1.iloc[j,i]-df01.iloc[j,i]
df1 = df1.drop(df1.columns[-1], axis=1)
#normalised the 1st iteration
df1.head()

df2=pd.read_csv('12.csv')
df02=pd.read_csv('02.csv')
for col in df2.columns:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')
for col in df02.columns:
    df02[col] = pd.to_numeric(df02[col], errors='coerce')
for i in range(0,250):
  for j in range(0,500):
    df2.iloc[j,i]=df2.iloc[j,i]-df02.iloc[j,i]
df2 = df2.drop(df2.columns[-1], axis=1)
#normalised 2nd iteration
df2.head()

df3=pd.read_csv('13.csv')
df03=pd.read_csv('03.csv')
for col in df3.columns:
    df3[col] = pd.to_numeric(df3[col], errors='coerce')
for col in df03.columns:
    df03[col] = pd.to_numeric(df03[col], errors='coerce')
for i in range(0,250):
  for j in range(0,500):
    df3.iloc[j,i]=df3.iloc[j,i]-df03.iloc[j,i]
df3 = df3.drop(df3.columns[-1], axis=1)
#normalised 3rd iteration
df3.head()

#now we have the datasets ready for plotting
#just find the medians and store in a numpy array as before
y=np.empty((250,3))#250 medians across 3 iterations
median1=df1.median();
y[:,0]=median1
median2=df2.median();
y[:,1]=median2
median3=df3.median();
y[:,2]=median3

x=np.empty(250,)
k=0
for i in range(50000,300000,1000):
  x[k]=i
  k+=1

plt.plot(x,y[:,0],linestyle='-',marker='.',linewidth=1,label='1Iteration1')
plt.plot(x,y[:,1],linestyle='-.',marker='*',linewidth=2,label='1Iteration2')
plt.plot(x,y[:,2],linestyle=':',marker='^',linewidth=3,label='1Iteration3')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
